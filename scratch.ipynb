```r
zs = c()
us = c()
for (i in seq(0,1,length=1000)) {
    zi = rnorm(1, mean=0.34, sd=0.3)
    q  = 1.4 * dnorm(zi, mean=0.34, sd=0.3)
    u  = runif(1, min=0, max=q)
    zs = append(zs, zi)
    us = append(us, u)
}
df=data.frame(x=zs, y=us)
```

```r
zs = c()
us = c()
for (i in seq(0,1,length=5000)) {
    zi = rnorm(1, mean=0.34, sd=0.3)
    q  = 1.4 * dnorm(zi, mean=0.34, sd=0.3)
    u  = runif(1, min=0, max=q)
    if (u > dbeta(zi, 2, 3))
        next
    zs = append(zs, zi)
    us = append(us, u)
}
betadf=data.frame(x=zs, y=us)
```

```r
ggplot(betadf, aes(x=x,y=y),color=variable)+geom_point(aes(col='(x,u)'))+geom_point(data=df, aes(y=norm, col='norm'))+geom_point(data=df, aes(y=beta, col='beta'))
```

```r
square = function(p, width) {
    x = p[1]
    y = p[2]
    lowx = x - width/2
    highx = x + width/2
    lowy = y - width/2
    highy = y + width/2
    if (lowy < 0) {
        stop("wrong y")
    }
    if (highy > 1.4*dnorm(lowx, mean=0.34, sd=0.3)) {
        stop("wrong y")
    }
    if (highy > 1.4*dnorm(highx, mean=0.34, sd=0.3)) {
        stop("wrong y")
    }
    return(c(lowx,highx,lowy,highy))
}
```

```r
square(c(0.5, 0.5), 0.2)
```

```r
in_square_F = function(r,square_) {
    in_square = function(r) {
        if (r['x'] >= square_[1] && r['x'] <= square_[2] && r['y'] >= square_[3] && r['y'] <= square_[4]) {
            return(T)
        }
        return(F)
    }
    return(in_square)
}
```

```r
ptilde = function(x) {
    p = 0.3 * dnorm(x, mean=4, sd=1) + 0.7 * dnorm(x, mean=7, sd=0.5)
    return(p)
}
```

```r
x=seq(0,10,length=1000)
df = data.frame(x=x, y=ptilde(x), norm=dnorm(x, mean=5, sd=2))
ggplot(df, color=variable)+geom_point(aes(x,y, col='p(x)'))+geom_point(aes(x,y=norm, col='q(x)'))+geom_point(aes(x,y=x/50,col='f(x)'))
```

```r
n=10000
sample.x = rnorm(n, mean=5, sd=2)
r.l = ptilde(sample.x) / dnorm(sample.x, mean=5, sd=2)
w.l = r.l/sum(r.l)
exp.x = sum(w.l * (sample.x/50))
exp.x
```
